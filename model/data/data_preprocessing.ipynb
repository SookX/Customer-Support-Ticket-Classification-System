```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./data/customer_support_tickets.csv')
```

```python
data
```

## Ticket Subject

```python
ts_vc = data['Ticket Subject'].value_counts()
ts_c = data['Ticket Subject'].unique()
```

```python
ts_c
```

```python
ts_dict = {}
for class_ in ts_c:
    ts_dict[class_] = ts_vc[class_]
```

```python
ts_dict
```

```python
plt.figure(figsize=(10,6))
plt.bar(ts_dict.keys(), ts_dict.values(), color='skyblue')
plt.xlabel('Issue Type')
plt.ylabel('Number of Complaints')
plt.title('Number of Complaints by Issue Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
classes = ['Product Purchased', 'Ticket Subject', 'Ticket Description']
```

```python
for class_ in classes:
    print(f"{class_} (NaN): {data[class_].isna().sum()}")
```

```python
preprocessed_data = data[classes]
preprocessed_data
```

```python
preprocessed_data
```

```python
def print_info(data, classes, index):
    piece = data.iloc[index]
    for class_ in classes:
        print(f"{class_} for index {index}: {piece[class_]}")

index = 1
print_info(preprocessed_data, classes, index)
```

```python
preprocessed_data.loc[:, 'Ticket Description'] = preprocessed_data['Ticket Description'].str.replace('\n', '', regex=False)
```

```python
index = 1
print_info(preprocessed_data, classes, index)
```

```python
preprocessed_data.loc[:, 'Ticket Description'] = preprocessed_data.apply(lambda row: row['Ticket Description'].replace('{product_purchased}', row['Product Purchased']), axis=1)
```

```python
index = 683
print_info(preprocessed_data, classes, index)
```

```python
preprocessed_data
```

```python
final_data = preprocessed_data.drop('Product Purchased', axis=1)
final_data
```

```python
final_data['Ticket Description']
```

```python
longest_string = final_data['Ticket Description'].str.len().idxmax()
longest_ticket_description = final_data.loc[longest_string, 'Ticket Description']
print(f"The longest string is: {len(longest_ticket_description)}")
```

```python
final_data.to_csv('./data/customer_support_tickets_preprocessed.csv', index=0)
```
